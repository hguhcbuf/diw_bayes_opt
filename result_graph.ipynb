```python
import json
params = []
target = []
data_str = ["/Users/jonghyun/Desktop/logs/cnn_guieded_bayes_go_copy_80.json"]

if isinstance(data_str, str):
    data_str = [data_str]

for data in data_str:
    with open(data, "r") as f:
        while True:
                try:
                    iteration = next(f)
                except StopIteration:
                    break

                iteration = json.loads(iteration)
                try:
                    
                    params.append(iteration["params"]),
                    target.append(iteration["target"]),
                    
                except KeyError:
                    pass
print(params)
print(target)
```
